```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("facebook/opt-125m")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
df_train = pd.read_csv("/content/drive/MyDrive/NLP_Papers/Data/multiclass_train.csv")
train, val = train_test_split(df_train, test_size=0.2, random_state=42, shuffle=True)
# df_train = pd.read_csv("/content/drive/MyDrive/NLP_Papers/Data/multiclass_train.csv") 
# df_valid = pd.read_csv("/content/drive/MyDrive/NLP_Papers/Data/multiclass_validation.csv")
# df_test = pd.read_csv("/content/drive/MyDrive/NLP_Papers/Data/multiclass_test.csv")

train.to_csv('/content/drive/MyDrive/NLP_Papers/Data/train_multiclass.csv', index=False)
val.to_csv('/content/drive/MyDrive/NLP_Papers/Data/val_multiclass.csv', index=False)
train_path='/content/drive/MyDrive/NLP_Papers/Data/train_multiclass.csv'
val_path='/content/drive/MyDrive/NLP_Papers/Data/val_multiclass.csv'

from datasets import load_dataset
train_dataset=load_dataset("csv", data_files=train_path)
val_dataset=load_dataset("csv", data_files=val_path)
test_dataset=load_dataset("csv", data_files="/content/drive/MyDrive/NLP_Papers/Data/multiclass_test.csv")
```

```python
def preprocess_function(examples):
   return tokenizer(examples["RomanUrduText"], truncation=True)
```

```python
tokenized_train = train_dataset.map(preprocess_function, batched=True)
tokenized_test = test_dataset.map(preprocess_function, batched=True)
tokenized_val = val_dataset.map(preprocess_function, batched=True)
```

```python
tokenized_train
```

```python
train_dataset = tokenized_train.map(lambda x: {"text": x["RomanUrduText"], "labels": x["Label"]})
test_dataset = tokenized_test.map(lambda x: {"text": x["RomanUrduText"], "labels": x["Label"]})
val_dataset = tokenized_val.map(lambda x: {"text": x["RomanUrduText"], "labels": x["Label"]})
```

```python
train_dataset["train"]["text"]
```

```python
from transformers import DataCollatorWithPadding
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
from transformers import AutoModelForSequenceClassification
num_label = 5
model = AutoModelForSequenceClassification.from_pretrained("facebook/opt-125m", num_labels=num_label)
```

```python
#compute 4
import evaluate
import numpy as np
from evaluate import load


f1_metric = evaluate.load("f1")
accuracy_metric = evaluate.load("accuracy")
precision_metric = evaluate.load("precision")
recall_metric = evaluate.load("recall")

def compute_metrics (eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis = -1)


    results = {}
    results.update(f1_metric.compute(predictions=preds, references = labels, average="micro"))
    results.update(accuracy_metric.compute(predictions=preds, references=labels))
    results.update(recall_metric.compute(predictions=preds, references = labels, average="micro"))
    results.update(precision_metric.compute(predictions=preds, references=labels, average="micro"))
    return results
```

```python
from transformers import TrainingArguments, Trainer
 
 
training_args = TrainingArguments(
   output_dir="/content/drive/MyDrive/NLP_Papers/Data",
   learning_rate=2e-5,
   per_device_train_batch_size=16,
   per_device_eval_batch_size=16,
   num_train_epochs=4,
   weight_decay=0.01,
   save_strategy="epoch",
)
 
trainer = Trainer(
   model=model,
   args=training_args,
   train_dataset=train_dataset['train'],
   eval_dataset=val_dataset['train'],
   tokenizer=tokenizer,
   data_collator=data_collator,
   compute_metrics=compute_metrics,
)
```

```python
#epoch
trainer.train()
```

```python
#epoch 4
trainer.evaluate()
```

```python
#prediction

y_test_predict = trainer.predict(test_dataset['train'])

# Take a look at the predictions
y_test_predict
```
